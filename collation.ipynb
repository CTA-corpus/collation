```python
#import libraries
from collatex import *
#import json
#import nltk #natural language toolkit; kit de processamento
#nltk.download('punkt')
import pandas as pd
```

```python
#data input
df1 = pd.read_excel("VMSSB-G1_wp.xlsx") #editar nome do ficheiro
df2 = pd.read_excel("VMSSB-G2_wp.xlsx")
df3 = pd.read_excel("VMSSB-E_wp.xlsx")
df4 = pd.read_excel("VMSSB-P_wp.xlsx")
len(df2.columns)
```

```python
 #atribuição de títulos às colunas
df1.columns=["id","w","lemma","c", "f", "nil"]
df2.columns=["id","w","lemma","c", "f", "nil"]
df3.columns=["id","w","lemma","c", "f", "nil"]
df4.columns=["id","w","lemma","c", "f", "nil"]
df3.head()
```

```python
#função para a tokenização
def get_tokens(df):
  tokens = []
  for id, w, lemma, c, f, nil in df.values:
    if lemma == "<unknown>": #regra para casos de lemma <unknown>
      tokens.append({"t": str(w)+" "})
    else:
      #tokens.append({"t": str(w)+" "}) #word
      #tokens.append({"t": str(w)+" ", "n": str(lemma)}) #word+lemma
      #tokens.append({"t": str(w)+" ", "n": str(lemma)+"_"+str(c)}) #word+lemma+cat
      tokens.append({"t": str(w)+" ", "n": str(lemma)+"_"+str(c)+":"+str(f)}) #word+lemma+cat+flex 
        #tokens.append({"t": str(w)+" ", "n": str(lemma)+"_"+str(tag)}) #sem separação de conteúdo
  return tokens

get_tokens(df3[:10])
```

```python
#colação com json input
collation = {"witnesses": [ {"id": "G1", "tokens": get_tokens(df1)}, {"id": "G2", "tokens": get_tokens(df2)}, {"id": "E", "tokens": get_tokens(df3)}, {"id": "P", "tokens": get_tokens(df4)}]}
```

```python
#visualização dos resultados
collate(collation, output="html2", layout="vertical", segmentation=False, near_match=True)
```
